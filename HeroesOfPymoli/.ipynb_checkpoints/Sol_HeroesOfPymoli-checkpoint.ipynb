```python
# Dependencies
import pandas as pd

# File to Load
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players

```python
# Find the total number of players

players_count = len(purchase_data['SN'].unique())

players_count_df = [{"Total Players" : players_count}]

# Convert it into a dataframe
players_count_df = pd.DataFrame(players_count_df)
players_count_df
```


## _Purchasing Analysis_


* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Find Average Price, Number of Purchases, Total Revenue

Number_of_Unique_Items = len(purchase_data['Item ID'].unique())
Average_price = purchase_data['Price'].mean()
Number_of_Purchases = len(purchase_data)
Total_Revenue = purchase_data['Price'].sum()
```

```python
# Create a Dataframe for the Summary

Summary = [{"Number of Unique Items" : Number_of_Unique_Items, 
            "Average Price" : Average_price,
            "Number of Purchases" : Number_of_Purchases,
            "Total Revenue" : Total_Revenue
           }]

Summary_df = pd.DataFrame(Summary)

# Use Map to format the "Average Price" & "Total Revenue" columns

Summary_df["Average Price"] = Summary_df["Average Price"].map("${:.2f}".format)
Summary_df["Total Revenue"] = Summary_df["Total Revenue"].map("${:.2f}".format)

#Display the summary data frame

Summary_df 
```

##  _Gender Demographics_


* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Get "Gender", "SN" columns from Purchase Data

Gender_df = purchase_data.loc[:,["Gender","SN"]].drop_duplicates()
Gender_df.head()
```

```python
# Find the count of each gender

Gender_count = Gender_df["Gender"].value_counts()

# Convert it into a DataFrame & Rename column name "Gender" to "Total Count"

Gender_Demographics_df = pd.DataFrame(Gender_count).rename(columns = {"Gender": "Total Count"})

# Find Percentage of each gender

Percentage_Players = Gender_Demographics_df["Total Count"]/Gender_Demographics_df["Total Count"].sum() * 100

# Convert it into a DataFrame
Gender_Demographics_df["Percentage of Players"] = Percentage_Players.map("{:.2f}%".format)

Gender_Demographics_df
```

## _Purchasing Analysis (Gender)_

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Find count of each gender

Gender_Purchase_Count = purchase_data["Gender"].value_counts()

# Creating Data Frame for Gender count

Gender_Analysis_df = pd.DataFrame(Gender_Purchase_Count).rename(columns= {"Gender" : "Purchase Count"})

Gender_Analysis_df
```

```python
#Find Average Purchase Price, Total Purchase Value, Avg Total Purchase per Person

Gender_Analysis_df["Average Purchase Price"] = purchase_data.groupby("Gender").mean()["Price"]
Gender_Analysis_df["Total Purchase Value"] = purchase_data.groupby("Gender").sum()["Price"]
Gender_Analysis_df["Avg Total Purchase per Person"] = Gender_Analysis_df["Total Purchase Value"]/Gender_Demographics_df["Total Count"]

# Use Map to format the required columns

Gender_Analysis_df["Average Purchase Price"] = Gender_Analysis_df["Average Purchase Price"].map("${:.2f}".format)
Gender_Analysis_df["Total Purchase Value"] = Gender_Analysis_df["Total Purchase Value"].map("${:.2f}".format)
Gender_Analysis_df["Avg Total Purchase per Person"] = Gender_Analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)

Gender_Analysis_df
```

## _Age Demographics_

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins for age groups
# Bins are 0,9,14,19,24,29,34,39,200

bins = [0,9,14,19,24,29,34,39,200]

# Create the names for the four bins

bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

Gender_df["Age Group"] = pd.cut(purchase_data["Age"],bins,labels = bin_names)

Gender_df.head()
```

```python
## Find the Value counts of the Age groups based off of the bins

Age_Demographics = Gender_df["Age Group"].value_counts().sort_index(axis=0,ascending=True)
Age_Demographics
```

```python
# Convert the data into a Data Frame

Age_Demographics_df = pd.DataFrame(Age_Demographics).rename(columns= {"Age Group" : "Total Count"})

# Finding Percentage of Players of each Age group

Percentage_of_Players = (Age_Demographics_df["Total Count"] / Age_Demographics_df["Total Count"].sum() *100)

#Add Percentage of Players to Age_Demographics DataFrame and use map to format

Age_Demographics_df["Percentage of Players"] = Percentage_of_Players.map("${:.2f}".format)

Age_Demographics_df
```

## _Purchasing Analysis (Age)_

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create bins for age groups
# Bins are 0,9,14,19,24,29,34,39,200

bins = [0,9,14,19,24,29,34,39,200]

# Create the names for the four bins

bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age Range"] = pd.cut(purchase_data["Age"],bins,labels = bin_names)

purchase_data
```

```python
# Create a DataFrame for Age analysis by takimng columns "Age Range","Purchase ID" from purchase data

Age_Analysis_df = purchase_data.loc[:,["Age Range","Purchase ID"]]

# Calculate Purchase Count, Avg Purchase Price, Total Purchase Value, Avg. Purchase Total per erson 

Age_Analysis_df = Age_Analysis_df.groupby(["Age Range"]).count().rename(columns = {"Purchase ID":"Purchase Count"})

Age_Analysis_df["Avg Purchase Price"] = purchase_data.groupby(["Age Range"]).mean()["Price"]

Age_Analysis_df["Total Purchase Value"] = purchase_data.groupby(["Age Range"]).sum()["Price"]

Age_Analysis_df["Avg Total Purchase per Person"] = (Age_Analysis_df["Total Purchase Value"]/Age_Demographics_df["Total Count"])

Age_Analysis_df
```

```python
# Using map to format Columns

Age_Analysis_df["Total Purchase Value"] = Age_Analysis_df["Total Purchase Value"].map("${:.2f}".format)
Age_Analysis_df["Avg Purchase Price"] = Age_Analysis_df["Avg Purchase Price"].map("${:.2f}".format)
Age_Analysis_df["Avg Total Purchase per Person"] = Age_Analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)
Age_Analysis_df
```

## _Top Spenders_

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Purchase_Analysis = purchase_data.loc[:,["SN","Price"]]
 
Purchase_Analysis
```

```python
# Calculate Purchase Count, Average Purchase Price, Total Purchase Value for each customer

Purchase_count = purchase_data["SN"].value_counts()
Total_Purchase_Value = purchase_data.groupby("SN").sum()["Price"]
Average_Purchase_Price = Total_Purchase_Value / Purchase_count


# Create a DataFrame Purchase_Analysis_df


Purchase_Analysis_df = pd.DataFrame({"Purchase Count":Purchase_count,
                                     "Average Purchase Price":Average_Purchase_Price,
                                     "Total Purchase Value": Total_Purchase_Value})


# Sort the table by Total Purchase Value

Top_Spenders_df = Purchase_Analysis_df.sort_values("Total Purchase Value", ascending = False)


# Use map to format the columns "Average Purchase Price", "Total Purchase Value"

Top_Spenders_df["Average Purchase Price"] = Top_Spenders_df["Average Purchase Price"].map("${:.2f}".format)
Top_Spenders_df["Total Purchase Value"] = Top_Spenders_df["Total Purchase Value"].map("${:.2f}".format)


# Display the Top 5 Spenders

Top_Spenders_df.head() 
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Display a preview of the summary data frame


```python
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value

Popular_items = purchase_data.groupby(["Item ID","Item Name"]).count()["Price"].rename("Purchase Count")
item_price = purchase_data.groupby(["Item ID","Item Name"]).mean()["Price"].rename("Item Price")
Total_Purchase_Value = purchase_data.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Purchase Value")


# Create DataFrame and sort by Purchase Count

Popular_items_df = pd.DataFrame({"Purchase Count" : Popular_items, 
                                 "Item Price" : item_price,
                                 "Total Purchase Value" : Total_Purchase_Value})

Popular_items_df.head()
```

```python
# Sort the DataFrame by Purchase Count

Most_Popular_items_df = Popular_items_df.sort_values("Purchase Count", ascending=False)

# Use map to format columns
Most_Popular_items_df["Item Price"] = Most_Popular_items_df["Item Price"].map("${:.2f}".format)
Most_Popular_items_df["Total Purchase Value"] = Most_Popular_items_df["Total Purchase Value"].map("${:.2f}".format)


# Display Top 5 Popular items

Most_Popular_items_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Display a preview of the data frame



```python
# Sort the above table by total purchase value in descending order

Most_Profitable_items_df = Popular_items_df.sort_values("Total Purchase Value", ascending = False)


# Use map to format columns

Most_Profitable_items_df["Item Price"] = Most_Profitable_items_df["Item Price"].map("${:.2f}".format)
Most_Profitable_items_df["Total Purchase Value"] = Most_Profitable_items_df["Total Purchase Value"].map("${:.2f}".format)

# Display Top 5 Profitable items

Most_Profitable_items_df.head()
```
